## 1. Histograma con la distribución de la población de la ciudad por edades

```r
# subset edades: nos quedamos con la 1ª fila (Bcn) y las columnas que representan la edad (de la columna 4 a la 104). 
df_edades <- demogr_bcn_data[1,c(4:104)]
head(df_edades)
```

```r
# miramos la edad mínima y la máxima
edad_min = names(df_edades[1]) # 0 
edad_max = names(df_edades[101]) # 100

# creamos un vector que contenga las edades
v_edades<-c(0:100)

# trasponemos el df edades 
df_edades_t<-t(df_edades)
# renombramos la columna
colnames(df_edades_t)[colnames(df_edades_t) == '1'] <- 'count'

# creamos el df final
df_distrib_pobl_bcn_edades <- data.frame("edad" = v_edades, df_edades_t)
# duplicamos las filas según el número de personas que tiene una determinada edad: si con 0 años hay 13633 personas, se crearán 13633 filas con un 0. 
df_distrib_pobl_bcn_edades_dup = uncount(df_distrib_pobl_bcn_edades, count, .remove = FALSE)
```

```r
# Creamos el histograma
hist(df_distrib_pobl_bcn_edades_dup$edad,
         main="Distribución de la población de Barcelona por edades",
         xlab="Edad", 
         border="white", 
         col="#7293cb")  
```

Podemos observar que este histograma presenta una cola derecha más larga, formando una asimetría negativa. Aunque como veremos en el siguiente apartado la media y la mediana son muy similares (entorno los 43 años), estas están ligeramente a la derecha de la moda (entorno a los 40 años). De los 0 a los 20 años se puede apreciar que la frecuencia varía muy poco, y es a partir de los 25 años que empieza a ascender. Los valores más frecuentes están entorno a los 35-45. Por tanto, en Barcelona, la mayoría de las personas tienen entre 35 y 45 años. A partir de este rango de edades, la frecuencia de estas empieza a disminuir a medida que la edad avanza (cuantos más años, menos gente), con la excepción de la edad 80, que presenta una frecuencia mucho menor. Este hecho podría ser debido a una baja natalidad en los años 40 del siglo pasado. 

## 2. Boxplot para la distribución de edades de la ciudad

- **Nota**: 
    - el boxplot realizado con la librería plotly no se puede visualizar en formato PDF pero sí en el entorno de Jupyter Notebook. Se han realizado dos boxplots, uno con plotly (interactivo) para cuando se visualice el código en Jupyter Notebook, y otro con ggplot2 para poder visualizarlo en el informe PDF. 

```r
# Boxplot interactivo usando plotly
bp <-plot_ly(
    data = df_distrib_pobl_bcn_edades_dup,
    y = ~edad,
    x = "Población de Barcelona",  
    #boxmean = TRUE, 
    boxmean = 'sd', # mostramos la media y la desviación estándar (línea discontinua)
    type = "box"
) %>% 
    layout(title = 'Distribución de edades de Barcelona¶')

bp
```

```r
# Boxplot usando ggplot
ggplot(df_distrib_pobl_bcn_edades_dup, aes(x='',y=edad)) + 
    geom_boxplot(fill="slateblue", alpha=0.4) +
    xlab("Población de Barcelona")
```

En este boxplot, que representa la distribución de edades de Barcelona, podemos observar varias cosas. El IQR no es muy grande, ya que la altura del boxplot va des de los 27 años (Q1) hasta los 61 años (Q3), es decir, la mayoría de las personas de Barcelona se encuentran entre este rango de edad. Por tanto, no hay una dispersión muy elevada: está bastante condensado. Esto significa que varía poco; es más consistente y fácil de predecir. Además, este boxplot es bastante simétrico ya que el Q2 está en el centro.

## 1. Verduras y hortalizas ricas en proteínas

Las proteínas son necesarias para poder construir el músculo después de un entrenamiento o competición. Es por ello que se ha seleccionado las verduras y hortalizas con más proteínas para que los deportistas las ingieran, por ejemplo, después de entrenar. 

```r
# df verduras y hortalizas
verd_hort_df <- data_fhv[data_fhv$Grupo == "Verduras y hortalizas",]
# df frutas
frutas_df <- data_fhv[data_fhv$Grupo == "Frutas",]
```

```r
# subset con el nombre de la verdura u hortaliza, el grupo y el valor proteico 
verd_hort_protein <- select(verd_hort_df, Nombre, Grupo, Proteina_g)
# ordenamos de manera descendente 
verd_hort_protein_desc <- arrange(verd_hort_protein, desc(Proteina_g))
# top 10 verduras y hortalizas ricas en proteínas
top_10_verd_hort_ricas_prot <- head(verd_hort_protein_desc,10)
top_10_verd_hort_ricas_prot
```

```r
# tamaño plot
options(repr.plot.width = 10, repr.plot.height = 5)
# barplot 
v_h_prot_p <- ggplot(data = top_10_verd_hort_ricas_prot, aes(x = Nombre, y = Proteina_g)) + 
                    geom_bar(stat = "identity", fill='burlywood3') +
                    labs(title = "Top 10 de verduras y hortalizas ricas en proteína", x = "Verduras y hortalizas", y = "Proteína (g por cada 100 g)")
# ggplotly(v_h_prot_p)
v_h_prot_p
```

Podemos observar que las hojas de parra son las hortalizas que mayor cantidad de proteínas tienen, seguidas del ajo, de la raíz de wasabi y del brécol. Por tanto, sería conveniente incluir hojas de parra, raíz de wasabi, brécol y coles de Bruselas en la dieta para favorecer la creación de músculo. Aunque el ajo haya salido en las primeras posiciones, este se debe consumir con moderación para evitar problemas intestinales. 

## 2. Verduras y hortalizas ricas en potasio

El potasio es un electrolito que participa en los procesos de relajación muscular, y su carencia puede producir debilidad muscular. Por tanto, los deportistas deben incluirlo en su dieta de manera regular. 

```r
# subset con el nombre de la verdura u hortaliza, el grupo y el potasio
verd_hort_potasio <- select(verd_hort_df, Nombre, Grupo, Potasio_mg)
# ordenamos de manera descendente 
verd_hort_potasio_desc <- arrange(verd_hort_potasio, desc(Potasio_mg))
# top 10 verduras y hortalizas ricas en potasio
top_10_verd_hort_ricas_potasio <- head(verd_hort_potasio_desc,10)
top_10_verd_hort_ricas_potasio
```

```r
# tamaño plot
options(repr.plot.width = 10, repr.plot.height = 5)
# barplot 
v_h_pot_p <- ggplot(data = top_10_verd_hort_ricas_potasio, aes(y = Nombre, x = Potasio_mg)) + 
                    geom_bar(stat = "identity", fill='darkorchid') +
                    labs(title = "Top 10 de verduras y hortalizas ricas en potasio", y = "Verduras y hortalizas", x = "Potasio (mg por cada 100 g)")
v_h_pot_p
```

Sería interesante que verduras fáciles de ingerir, como la patata, acelgas y las espinacas, fueran incluidas de manera regular en la dieta para garantizar unos niveles de potasio adecuados. 

## 3. Verduras y hortalizas ricas en fósforo

Micronutrientes como el fósforo es muy beneficioso en los deportistas porque ayuda al corazón, mejora la contractilidad del músculo, y participa en los procesos de creación de energía. Además, reduce el ácido láctico mientras se practica ejercicio.

```r
# subset con el nombre de la verdura u hortaliza, el grupo y el fósforo
verd_hort_fosforo <- select(verd_hort_df, Nombre, Grupo, Fosforo_mg)
# ordenamos de manera descendente 
verd_hort_fosforo_desc <- arrange(verd_hort_fosforo , desc(Fosforo_mg))
# top 10 verduras y hortalizas ricas en fósforo
top_10_verd_hort_ricas_fosforo <- head(verd_hort_fosforo_desc,10)
top_10_verd_hort_ricas_fosforo
```

```r
# tamaño plot
options(repr.plot.width = 10, repr.plot.height = 5)
# barplot 
v_h_pot_p <- ggplot(data = top_10_verd_hort_ricas_fosforo, aes(y = Nombre, x = Fosforo_mg)) + 
                    geom_bar(stat = "identity", fill='#a5d0c6', width=0.8) + 
                    labs(title = "Top 10 de verduras y hortalizas ricas en fósforo", y = "Verduras y hortalizas", x = "Fósforo (mg por cada 100 g)")
v_h_pot_p
```

Teniendo en cuenta lo anterior, es conveniente que deportistas ingieran alimentos ricos en fósforo como pueden ser las alcachofas frescas o el brécol.

## 4. Frutas, verduras y hortalizas con más agua

La deshidratación supone un grave problema para aquellos que la sufren. Los deportistas deben estar bien hidratados en todo momento para optimizar su rendimiento y no tener problemas de salud.

Por ello, es recomendable que, sobre todo en días de mucho calor, tanto antes como después de realizar el ejercicio físico se ingieran alimentos ricos en agua. Por ejemplo, calabacines, pepinos o frutas como la sandía y el melón.

Ingerir las suficientes calorías es necesario sobre todo para aquellos deportistas cuyo objetivo sea aumentar la masa muscular. Muchos de los alimentos con más calorías no son saludables, y es por ello por lo que se han seleccionado las frutas que contienen más kcal y pueden ayudar a llegar al objetivo en cuestión de una forma más saludable.

## 6. Verduras, hortalizas y frutas con más hidratos de carbono

Los hidratos de carbono son la base de la alimentación e ingerir las cantidades necesarias es fundamental para reponer los niveles de glucógeno sobre todo después de realizar actividad física intensa. 

Hoy en día, muchos hidratos de carbono se ingieren de alimentos preprocesados y poco saludables, como la bollería industrial y algunas harinas refinadas, que, entre otras cosas, pueden causar inflamación en el cuerpo. Esto es lo último que un deportista quiere, ya que después de realizar ejercicio físico el cuerpo ya está lo suficientemente inflamado. Una fuente de hidratos de carbono saludable son las verduras, hortalizas y frutas. Es por ello que se han seleccionado las 10 que contienen más carbohidratos.
